```python
import pandas as pd
import plotly.express as px
```

```python
data = {'Name': ['John', 'Mary', 'Bob', 'Sarah', 'Tom', 'Lisa'], 'Department': ['IT', 'Marketing', 'Sales', 'IT', 'Finance', 'Marketing'], 'Age': [30, 40, 25, 35, 45, 28], 'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'], 'Salary': [50000, 60000, 45000, 55000, 70000, 55000], 'Experience': [3, 7, 2, 5, 10, 4]}

employee_df = pd.DataFrame(data)

employee_df.head()
```

```python
fig = px.bar(employee_df, x="Department", y="Experience", color="Experience", height=200)
fig.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(employee_df["Department"], employee_df["Salary"])

plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Departmental Salary")

plt.show()
```

```python

import matplotlib.pyplot as plt

colors = ['red', 'green', 'blue', 'purple']

plt.bar(employee_df["Department"], employee_df["Salary"], color= colors)

plt.xlabel("Department")
plt.ylabel("Salary")
plt.title("Departmental Salary")

plt.show()
```

```python
fig = px.pie(employee_df, values='Salary', names='Department',
             color='Department', title='Departmental Salary')

fig.update_traces(textposition='inside', textinfo='percent+label')

fig.update_layout(uniformtext_minsize=14, uniformtext_mode='hide')

fig.show()
```

```python
plt.pie(employee_df["Salary"], labels=employee_df["Department"], autopct="%1.1f%%")

plt.axis("equal")  # Equal aspect ratio ensures a circular pie chart

plt.title("Departmental Salary Distribution")
plt.show()
```
