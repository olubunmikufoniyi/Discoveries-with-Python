```python

shopping_list = ["apples", "bread", "milk", "eggs", "cheese"]

while True:
    print("Shopping List Menu:")
    print("1. View Shopping List")
    print("2. Add Item to Shopping List")
    print("3. Remove Item from Shopping List")
    print("4. Exit")

    choice = input("Enter your choice (1, 2, 3, or 4): ")

    if choice == "1":
        print("Current Shopping List:", shopping_list)
    elif choice == "2":
        item_to_add = input("Enter the item to add: ")
        shopping_list.append(item_to_add)
        print(f"{item_to_add} added to the shopping list.")
    elif choice == "3":
        item_to_remove = input("Enter the item to remove: ")
        if item_to_remove in shopping_list:
            shopping_list.remove(item_to_remove)
            print(f"{item_to_remove} removed from the shopping list.")
        else:
            print(f"{item_to_remove} not found in the shopping list.")
    elif choice == "4":
        print("Exiting the menu.")
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")

shopping_list = ["apples", "bread", "milk", "eggs", "cheese"]

for item in shopping_list:
    print(item)

max_items = 5
shopping_list = []

while True:
    print("Shopping List Menu:")
    print("1. View Shopping List")
    print("2. Add Item to Shopping List")
    print("3. Remove Item from Shopping List")
    print("4. Exit")

    choice = input("Enter your choice (1, 2, 3, or 4): ")

    if choice == "1":
        print("Current Shopping List:", shopping_list)
    elif choice == "2":
        if len(shopping_list) < max_items:
            item_to_add = input("Enter the item to add: ")
            shopping_list.append(item_to_add)
            print(f"{item_to_add} added to the shopping list.")
        else:
            print(f"Maximum limit of {max_items} items reached. Cannot add more.")
    elif choice == "3":
        item_to_remove = input("Enter the item to remove: ")
        if item_to_remove in shopping_list:
            shopping_list.remove(item_to_remove)
            print(f"{item_to_remove} removed from the shopping list.")
        else:
            print(f"{item_to_remove} not found in the shopping list.")
    elif choice == "4":
        print("Exiting the menu.")
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
# List for general shopping items
shopping_list = ["apples", "bread", "milk", "eggs", "cheese"]

# Tuple for prices corresponding to each item
item_prices = ("apples", 1.5), ("bread", 2.0), ("milk", 3.0), ("eggs", 2.5), ("cheese", 4.0)

# Set for items on sale
sale_items = {"bread", "milk"}

# Dictionary for quantities of each item
item_quantities = {"apples": 3, "bread": 2, "milk": 1, "eggs": 2, "cheese": 1}

# Display the original shopping list
print("Original Shopping List:", shopping_list)

# Add an item to the shopping list
shopping_list.append("orange")

# Display the updated shopping list
print("Updated Shopping List:", shopping_list)

# Access the price of bread from the tuple
bread_price = next(price for item, price in item_prices if item == "bread")
print("Price of bread:", bread_price)

# Check for items on sale
on_sale_items = shopping_list.intersection(sale_items)
print("Items on Sale:", on_sale_items)

# Increase the quantity of eggs
item_quantities["eggs"] += 1
print("Updated Quantities:", item_quantities)
```
