```python
import pandas as pd

url = 'https://drive.google.com/file/d/1I8eV4-8p61CNNlVJzzho2xeoZ5-P7Q0F/view'
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)

```

```python
df.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
from datetime import datetime

date_format = "%Y%m%d %H%M%S"

df["DATE"] = pd.to_datetime(df["DATE"], format=date_format)
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Year'] = pd.DatetimeIndex(df['DATE']).year
```

```python
df.head()
```

```python
df['Year'].unique()
```

```python
# Filter data for Tunisia and Cameroon
tunisia_data = df[df['COUNTRY'] == 'Tunisia']
cameroon_data = df[df['COUNTRY'] == 'Cameroon']

# Group data by date and calculate average temperature for each day
tunisia_avg_temp = tunisia_data.groupby('Year')['TAVG'].mean()
cameroon_avg_temp = cameroon_data.groupby('Year')['TAVG'].mean()
```

```python
# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(tunisia_avg_temp.index, tunisia_avg_temp.values, label='Tunisia', marker='o')
plt.plot(cameroon_avg_temp.index, cameroon_avg_temp.values, label='Cameroon', marker='o')
plt.title('Average Temperature Fluctuations in Tunisia and Cameroon')
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent cropping
plt.show()
```

```python
# Filter data for Senegal
senegal_data = df[df['COUNTRY'] == 'Senegal']
```

```python
senegal_data.head()
```

```python
senegal_1980_2000 = senegal_data[(senegal_data['DATE'] >= '1980') & (senegal_data['DATE'] <= '2000')]
```

```python
senegal_2000_2023 = senegal_data[(senegal_data['DATE'] >= '2000') & (senegal_data['DATE'] <= '2023')]
```

```python
import matplotlib.pyplot as plt
#Create subplots
fig, ax = plt.subplots()

# Plot histograms for the two time periods using Pandas plot.hist()
senegal_1980_2000.plot.hist(ax=ax, bins=20, alpha=0.5, label='1980-2000')
senegal_2000_2023.plot.hist(ax=ax, bins=20, alpha=0.5, label='2000-2023')

# Customize the plot
ax.set_xlabel('Temperature')
ax.set_ylabel('Frequency')
ax.set_title('Temperature Distribution in Senegal')
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot histograms
plt.figure(figsize=(10, 6))

plt.hist(senegal_1980_2000['TAVG'], bins=15, alpha=0.5, label='1980-2000')
plt.hist(senegal_2000_2023['TAVG'], bins=15, alpha=0.5, label='2000-2023')

plt.title('Temperature Distribution in Senegal')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

The above histogram of Senegal is a spread of tempretures ranging from 1980 to 2023. The histogtam in blue shows the distribution of temperatures from the years 1980 and 2020 while the brown histogram shows results for the years 2020 to 2023.
