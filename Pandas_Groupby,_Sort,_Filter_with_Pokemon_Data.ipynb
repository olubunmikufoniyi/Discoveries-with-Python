```python
import pandas as pd
df = pd.read_csv('pokemon_data.csv')
print(df.head(5))
```

Reading Data in Pandas

```python
##Read Headers
print(df.columns)
```

```python
##Read each Column
print(df['Name'])
```

```python
print(df[['Name','Type 1','HP'][0:5]])
```

```python
df.head()
```

```python
poke_table = df.pivot_table(values='HP', index='Type 1',aggfunc='mean')
print(poke_table)
```

```python
poke_table.plot(kind="bar")
```

```python
poke_table.plot(kind='line')
```

```python
poke_table.describe()
```

Filtering Data Frame

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] =='Poison') & (df['HP'] > 70)]
new_df.reset_index(drop=True,inplace=True)
```

```python
new_df.head()
```
